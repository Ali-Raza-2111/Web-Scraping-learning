```python
import pandas as pd

# Create the new DataFrame
data_new = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam', 'Headphones', 'Speaker'],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Electronics', 'Audio', 'Audio'],
    'Price_USD': [1200, 25, 75, 300, 50, 150, 100],
    'Stock_Quantity': [50, 200, 150, 30, 100, 80, 60],
    'Rating': [4.5, 4.0, 4.2, 4.7, 3.8, 4.3, 4.1]
}
df_products = pd.DataFrame(data_new)

print("Here's your new DataFrame:")
print(df_products)
print("\n")

# Practice Questions:
print("Practice Question 1:")
print("Select the 'Product' and 'Price_USD' columns.")
df_products[['Product','Price_USD']]
print("\nPractice Question 2:")
print("Select the 'Category', 'Stock_Quantity', and 'Rating' columns.")
df_products[['Category','Stock_Quantity','Rating']]
print("\nPractice Question 3:")
print("Select the 'Product', 'Category', and 'Price_USD' columns.")
df_products[['Product','Category','Price_USD']]
print("\nPractice Question 4:")
print("Select all columns except 'Rating' and 'Stock_Quantity'.")
df_result = df_products.drop(['Rating','Stock_Quantity'],axis=1)
print(df_result)
print("\nPractice Question 5:")
print("Select the 'Product' and 'Rating' columns, in that specific order.")
df_products[['Product','Rating']]
```
